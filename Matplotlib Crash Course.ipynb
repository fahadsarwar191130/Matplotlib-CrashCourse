```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### First Graph

```python

x = [1,2,3,4,5]
y = [2,4,6,8,10]

font = {'family': 'serif',
        'color':  'green',
        'weight': 'normal',
        'size': 16,
        }
plt.figure(figsize = [15,10], dpi = 300.0)

plt.plot(x,y, label = 'Run Rate',linestyle= '--', linewidth=1, marker= 'H', markeredgecolor = 'blue',  markersize=18, color = 'orange' )

x2 = np.arange(1,10,2)
plt.plot(x2[:6],x2[:6]**2)
plt.plot(x2[3:],x2[3:]**2, linestyle= '--')

plt.xlabel("Balls" , fontdict = font)
plt.ylabel("Overs" , fontdict = font)
plt.legend()

plt.savefig('graph.png', dpi = 500)

plt.show()


```

### Bar Charts

```python
x = ['Covid-19' , 'Heart Attack' , 'Heat Stroke']
y = [500, 300 , 650]

bars = plt.bar(x,y,0.5,20,align = 'center')

h = ['o', '+' , '*']

for i in bars:
    i.set_hatch(h.pop(0))

"""
bars[0].set_hatch('o')
bars[1].set_hatch('+')
bars[2].set_hatch('*')
"""

plt.show()
```

### Real World Examples

```python
gas = pd.read_csv('gas_prices.csv')
print('gas')

plt.plot(gas.Year, gas['Mexico'], 'r.-', label = 'Mexico')
plt.plot(gas.Year, gas['Australia'] , label = 'Aus')
plt.plot(gas.Year, gas['Germany'] , label = 'Ger')
plt.plot(gas.Year, gas['South Korea'] , label = 'SK')

plt.xticks(gas.Year[::2])
plt.title('Gas Prices (in USD)')

plt.xlabel('Price in USD')
plt.ylabel('Price')

plt.legend()

plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')
print('gas')

plt.figure(figsize = (15,10), dpi = 300)

for country in gas:
    if country != 'Year':
            plt.plot(gas.Year, gas[country], label = country)
        
plt.xticks(gas.Year[::2])
plt.title('Gas Prices (in USD)')

plt.xlabel('Price in USD')
plt.ylabel('Price')

plt.legend()
plt.savefig('gasgraph')
plt.show()
```
